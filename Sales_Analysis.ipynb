```python
# import python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#read CSV
sales = pd.read_csv(r"E:\Downloads\Python_Diwali_Sales_Analysis\Python_Diwali_Sales_Analysis\Diwali Sales Data.csv",encoding='unicode_escape')
```

```python
#delete NULL column
sales.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
#To remove NULL values from Coloumn
sales.dropna(inplace = True)
```

```python
#Data Type change for Amount
sales['Amount']=sales['Amount'].astype(int)
```

Exploratory Data Analysis 

```python
# plotting a bar chart for Gender and it's count
ax = sns.countplot(x='Gender',data=sales)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Order Count By Gender",fontsize=14)
```

```python
#Gender wise Sales
ax = sales.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = True)
ax =sns.barplot(x='Gender',y='Amount',data=ax)
for bars in ax.containers:
    ax.bar_label(bars,fmt='%.0f')
plt.title("Sales By Gender",fontsize=14)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

AGE

```python
#plotting a bar chart for Age_group and it's count
ax=sns.countplot(data = sales, x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Order Count By Age Group",fontsize=14)
```

```python
#Age group wise Sales
ax = sales.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
sns.set({'figure.figsize':(15,5)})
ax = sns.barplot(x='Age Group',y='Amount',data=ax)
for bars in ax.containers:
    ax.bar_label(bars,fmt='%d')
plt.title("Sales By Age Group",fontsize=14)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

States

```python
# plotting a bar chart for State 
ax = sns.countplot(data=sales,x='State')
sns.set({'figure.figsize':(30,20)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Order Count By State",fontsize=14)
```

```python
#Total Numbers of orders State wise
ax=sales.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)
sns.set({'figure.figsize':(19,5)})
ax=sns.barplot(x='State',y='Orders',data=ax)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Top 10 States By Order Count",fontsize=14)
```

```python
# State wise Total Sales
ax = sales.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set({'figure.figsize':(15,5)})
ax=sns.barplot(x='State',y='Amount', data = ax)
for bars in ax.containers:
    ax.bar_label(bars,fmt='%d')
plt.title("Top 10 States By Sales",fontsize=14)
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

Marital Status

```python
#Marital Status wise Order Count
sns.set({'figure.figsize':(15,5)})
ax = sns.countplot(data=sales,x='Marital_Status',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Order Count By Marital Status",fontsize=14)
```

```python
#Marital and Gender wise Amount
ax = sales.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.set({'figure.figsize':(5,5)})
ax =sns.barplot(x='Marital_Status',y='Amount',data=ax,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars,fmt='%d')
plt.title("Sales By Marital Status",fontsize=14)
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

Occupations

```python
# Occupation wise Count
sns.set({'figure.figsize':(15,5)})
ax=sns.countplot(data=sales, x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Order Count By Occupations",fontsize=14)
```

```python
#Group by Occ and Amount val
ax = sales.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set({'figure.figsize':(25,5)})
ax =sns.barplot(x='Occupation',y='Amount',data =ax)
for bars in ax.containers:
    ax.bar_label(bars,fmt='%d')
plt.title("Sales By Occupations",fontsize=14)
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

Product Category

```python
#Product Category Wise Orderd
ax=sales.groupby('Product_Category',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
ax=sns.barplot(data=ax,x='Product_Category',y='Orders')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Top 10 Product Category By Orders Count",fontsize=14)
```

```python
#Categoty wise top 10 Count of Orders
ax=sales.groupby(['Product_Category'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending =False).head(10)
sns.set({'figure.figsize':(15,10)})
ax=sns.barplot(x='Product_Category',y='Orders',data=ax)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Top 10 Product Category by Sales",fontsize=14)
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*
